```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Прочитайте файл 2017_jun_final.csv за допомогою методу read_csv

file_id = '1JMYqXipZpz9Y5-vyxvLEO2Y1sRBxqu-U'
url = f'https://drive.google.com/uc?id={file_id}'
data = pd.read_csv(url)
```

```python
#Прочитайте отриману таблицю, використовуючи метод head

data.head()
```

```python
#Визначте розмір таблиці за допомогою методу shape

data.shape
```

```python
#Визначте типи всіх стовпців за допомогою dataframe.dtypes

data.dtypes
```

```python
#Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

pd.isnull(data).sum()
```

```python
#Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"

data_1 = ['Язык.программирования'] + list(data.dropna(axis=1, how='any').columns)
data = data[data_1]
```

```python
#Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець "Мова.програмування"

pd.isnull(data).sum()
```

```python
#Видаліть усі рядки у вихідній таблиці за допомогою методу dropna

data.dropna()
```

```python
#Визначте новий розмір таблиці за допомогою методу shape

data.shape
```

```python
#Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python

python_data = data[data['Язык.программирования'] == 'Python']
```

```python
#Визначте розмір таблиці python_data за допомогою методу shape

python_data.shape
```

```python
#Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"
python_data = python_data.groupby('Должность')
```

```python
#Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть
#мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

df_1 = python_data['Зарплата.в.месяц'].agg(['min', 'max'])
df_1
```

```python
#Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

def fill_avg_salary(data):
  return (data['min']+data['max'])/2

avg = df_1.apply(fill_avg_salary, axis=1)
df_1['avg'] = avg
df_1
```

```python
#Створіть описову статистику за допомогою методу describe для нового стовпчика.

df_1['avg'].describe()
```

```python
#Збережіть отриману таблицю в CSV файл

df_1.to_csv('2017_jun_final_2.csv')
```

```python
#Залежність рівня зарплати від рівня володіння англійською мовою (Для Python)
python_data = data[data['Язык.программирования'] == 'Python']
sns.boxplot(x='Уровень.английского', y='Зарплата.в.месяц', hue='Должность', data=python_data)
```

```python
#Залежність зарплати від посади, окремі графіки для жінок та чоловіків
sns.set_context('paper')
gfg = sns.relplot(x='Зарплата.в.месяц', y='Должность', kind='scatter', hue='Пол', col='Пол', data=data)
gfg.set(xlabel ="Зарплата", ylabel = "Посада",title ='Залежність зарплати від посади')
```

```python
data_2 = data.groupby('Город')['Зарплата.в.месяц'].agg('mean')
data_2.plot(kind='bar', color=['r', 'b', 'g', 'y'], hatch='//')
plt.xlabel('Місто')
plt.ylabel('Зарплата')
plt.title('Середня зарплата в залежності від міста')
plt.show()
```
